## A Brief Canadian History of GIS

-   1951 - Census shows rapid urbanization

-   1957 - Canadian Senate investigates the "Land Use Problem in Canada"

-   1963 - Canada Land Inventory established, work on CGIS begins

-   1970s - 12,000 map sheets are produced, the first digital self-portrait of a nation

## AWS GIS Data Resources

To access public AWS data, first set environment variable in QGIS through the Python Console (Plugins \> Python Console):


```{import os}
os.environ['AWS_NO_SIGN_REQUEST'] = 'YES'
```


## British Columbia GIS Data Resources

[LidarBC Portal](https://lidar.gov.bc.ca/pages/download-discovery)

[LidarBC ArcGIS Feature Server](https://services6.arcgis.com/ubm4tcTYICKBpist/ArcGIS/rest/services/LiDAR_BC_S3_Public/FeatureServer)

[BC Data Catalogue](https://catalogue.data.gov.bc.ca/)

[BC Data Catalogue Web Feature Server (WFS)](http://openmaps.gov.bc.ca/geo/pub/wfs?request=GetCapabilities)

## UBC GIS Data Resources

[MGEM Data Store](https://206-12-122-94.cloud.computecanada.ca/)

[UBC Campus and Community Planning](https://github.com/UBCGeodata/ubc-geospatial-opendata)

[UBC ArcGIS Online](https://ubc.maps.arcgis.com/home/index.html)

## Open Educational Resources at UBC

[Geomatics for Environmental Management: An Open Textbook for Students and Practitioners](https://opengeomatics.ca)

[UBC Geomatics Community of Practice](https://cop.opengeomatics.ca)

## GIS Coursework in the Faculty of Forestry
